```python
import plotly.express as px
import pandas as pd
import glob
import os
```

```python
list_of_files = glob.glob('/Users/ralph/Documents/python_dashboard/california_covid_data_dashboard_project/covid19*.csv')
latest_file = max(list_of_files, key=os.path.getctime)
print(latest_file.split("\\")[-1])

df = pd.read_csv(latest_file.split("\\")[-1])
```

```python
df_t=df.T

def clean_index(s):
    s = s.replace("/", "-")
    return s

df_t.index = df_t.index.map(clean_index)
df = df.drop(columns=['area_type', 'population', 'cases', 'deaths', 'cumulative_deaths', 'total_tests'])
df = df.drop(columns=['cumulative_total_tests', 'positive_tests', 'cumulative_positive_tests', 'reported_cases'])
df = df.drop(columns=['cumulative_reported_cases', 'reported_deaths', 'cumulative_reported_deaths', 'reported_tests'])
```

```python
# initialize df with three columns: Date, Case Count, and County
case_column = df.pop("cumulative_cases")
df.insert(1, "Case Count", case_column)
df = df.rename(columns={"date": "Date", "area": "County"})

# save long df for dash app
df.to_csv("time_series_plotly.csv")
```

```python
fig = px.scatter(df, x='Date', y='Case Count', color='County')
fig.update_traces(mode='markers+lines')
fig.show()
```
